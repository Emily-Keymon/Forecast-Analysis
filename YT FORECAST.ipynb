## Extract/Transform Data

```python
#Import libraries
import pandas as pd
```

```python
# Load JAX OPEN ORDERS from Resources folder
file = "Resources/JAX OPEN ORDERS.xlsx"

# Read excel file into DataFrame, set ITEM as index
jax_data = pd.read_excel(file, index_col="Item")

#Check dataframe
jax_data.head()
```

```python
# Check count
jax_data.count()
```

```python
# Add ORG column
jax_data["ORG"] = "JAX"
jax_data.head()
```

```python
# Load LEX OPEN ORDERS from Resources folder
file = "Resources/LEX OPEN ORDERS.xlsx"

# Read excel file into DataFrame, set ITEM as index
lex_data = pd.read_excel(file, index_col="Item")

#Check dataframe
lex_data.head()
```

```python
# Check count
lex_data.count()
```

```python
# Add ORG column
lex_data["ORG"] = "LEX"
lex_data.head()
```

```python
# Merge jax_data and lex_data
both = [jax_data, lex_data]
total_data = pd.concat(both)
total_data.head()
```

```python
# Check count
total_data.count()
```

```python
# Check data types
total_data.dtypes
```

```python
# Change columns (take out spaces for future use)
total_data = total_data.rename(columns = {"Open Qty": "Open_Qty",
                                          "Extended Price": "Extended_Price",
                                          "Item Cost": "Item_Cost",
                                          "Material Cost": "Material_Cost",
                                          "Material OH Cost": "Material_OH_Cost",
                                          "Resource Cost": "Resource_Cost",
                                          "Overhead Cost": "Overhead_Cost",
                                          "OSP Cost": "OSP_Cost"})
total_data.head()
```

```python
# Check data types
total_data.dtypes
```

```python
# Extended Cost calculation
total_data["Extended_Cost"] = total_data.Open_Qty * total_data.Item_Cost.round(2)
total_data
```

```python
# Material calculation (Open Qty * Material Cost)
total_data["Material"] = total_data.Open_Qty * total_data.Material_Cost.round(2)
total_data.head()
```

```python
# Labor calculation (Open Qty * Resource Cost)
total_data["Labor"] = total_data.Open_Qty * total_data.Resource_Cost.round(2)
total_data.head()
```

```python
# Overhead calculation (Open Qty * Overhead Cost)
total_data["Overhead"] = total_data.Open_Qty * total_data.Overhead_Cost.round(2)
total_data.head()
```

```python
# Outside processing calculation (Open Qty * OSP Cost)
total_data["OSP"] = total_data.Open_Qty * total_data.OSP_Cost.round(2)
total_data.head()
```

```python
# Calculate COGS (Material Cost + Labor + Overhead Cost + Outside Processing (OSP Cost))
#total_data["COGS"] = total_data.Material_Cost + total_data.Overhead_Cost + total_data.OSP_Cost
total_data["Total"] = total_data.Material_Cost.round(2) + total_data.Overhead_Cost.round(2)
total_data
```

```python
# Check data types
total_data.dtypes
```

```python
# Calculate Gross Margin Dollars (Sales - COGS:  Extended Price - Extended Cost)
total_data["Margin"] = total_data.Extended_Price.round(2) - total_data.Extended_Cost.round(2)
total_data.head()
```

```python
# Calculate Gross Margin Percent (Margin / Extended Price)
total_data["Margin%"] = total_data.Margin.round(2) / total_data.Extended_Price.round(2)
total_data.head()
```
